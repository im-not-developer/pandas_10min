```python
import pandas as pd
```

```python
import numpy as np
df = pd.DataFrame(
            {
              "a" : [4, 5, 6, 6, np.nan],
              "b" : [7, 8, np.nan, 9, 9],
              "c" : [10, 11, 12,np.nan, 12]
            },
            index = pd.MultiIndex.from_tuples(
              [('d', 1),('d', 2),('e', 2),('e', 3),('e', 4)], 
              names=['n', 'v'])
            )
df
```

```python
#기본값이 5
df.head()
```

```python
"""
samaple은 특정 비율로 데이터를 불러옴(랜덤하게 가져옴)
frac 의 값은 0.1 부터 1까지 지정을 할 수 있음

"""
df.sample(frac=1)
```

```python
"""
sample에는 다양한 매개변수가 있다.
n 매개변수가 들어감 

Randomly select n rows.

"""
df.sample(n=2)
```

```python

"""
시작하는 인덱스와 끝나는 인덱스를 정해줄 수 있다.

값이 안나오는 이유는 해당 인덱스가 없기 때문이다.
"""

df.iloc[10:20]
```

```python
"""
인덱스 1부터 3까지 가져온다.
"""
df.iloc[1:3]
```

```python
"""
인덱스 0부터 3까지
"""

df.iloc[:3]
```

```python
"""
인덱스 3부터 끝까지
"""

df.iloc[3:]
```

```python
"""
마지막 인덱스 -2부터 끝까지 가져온다.
"""
df.iloc[-2:]
```

```python
df.iloc[:-2]
```

```python
df = pd.DataFrame({
    'a': [1,10,8,11,-1],
    'b': list('abdce'),
    'c': [1.0,2.0,np.nan,3.0, 4.0]
})
df
```

```python
# 가장 큰값
df.nlargest(3, 'a')
```

```python

df.nlargest(1, 'a')
```
