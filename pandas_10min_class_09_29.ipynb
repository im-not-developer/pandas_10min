# Handling Missing Data
## **fillna, dropna**


* filna 데이터를 채우는 것 
* dropna 데이터를 버리는 것


```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame([[np.nan, 2, np.nan, 0], [3, 4, np.nan, 1],
                   [np.nan, np.nan, np.nan, 5]],
                  columns=list('ABCD'))

df
```

```python
#  dropna 데이터를 버리는 것
"""
  axis 는 1 컬럼을 의미 , 0 로우를 의미
  how 는 전부 null 일 때 drop을 시킨다.
"""
df.dropna(axis=1, how='all')
```

```python
"""
how='any'는 하나라도 null 값이 있다면 drop을 시킨다.
"""
df.dropna(axis=1, how='any')
```

```python
df.dropna(axis=0, how='all')
```

```python
df.dropna(axis=0, how='any')
```

```python
df.fillna(0)
```

```python
values = {'A': 0, 'B': 1, 'C': 2, 'D': 3}
values
```

```python
df.fillna(value=values)
```

```python
df['D'].mean()
```

```python
fill_na_value = df['D'].max()
fill_na_value
```

```python
df.fillna(fill_na_value)
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df.notnull().sum()
```
