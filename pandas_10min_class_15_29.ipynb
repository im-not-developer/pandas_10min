# Group Data
**groupby 활용하여 다양한 데이터 집계를 활용하기**

```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('mpg')
df.head()
```

```python
"""
origin 기준으로 그룹화를 하고 사이즈를 확인
"""
df.groupby(by='origin').size()
```

```python
"""
위 와 같이 사이즈를 구하는 방법
"""
df['origin'].value_counts()
```

```python
df.groupby(by='origin')['cylinders'].size()
```

```python
pd.DataFrame(df.groupby(['origin', 'model_year'])['cylinders'].mean())
```

```python
df2 = pd.DataFrame(
    [[4, 7, 10],
     [5, 8, 11],
     [6, 9, 12]],
     index=[1, 2, 3],
     columns=['a', 'b', 'c'])
df2
```

```python
df2.shift(1)
```

```python
df2['b'].shift(1)
```

```python
df['model_year'].rank(method="min").value_counts()
```

```python
df['model_year'].rank(pct=True, method="first")
```
