특정 행만 뽑아서 데이터 프레임화를 한다.


```python
import pandas as pd
```

```python
df = pd.DataFrame(
            {
              "a" : [4, 5, 6],
              "b" : [7, 8, 9],
              "c" : [10, 11, 12]
            },
            index = pd.MultiIndex.from_tuples(
              [('d', 1),('d', 2),('e', 2)], 
              names=['n', 'v'])
            )
df
```

```python
"""
df에 length 라는 컬럼이 없기 때문에 에러가 남.
"""
df[df.length > 7]
```

```python
"""
a 컬럼에 7보다 큰 값이 없다.
"""
df[df.a > 7]
```

```python
df[df.a < 7]
```

```python
df.a > 7
```

```python
df[df['c'] > 7]
```

```python

# 중복된 값들을 추가
df = pd.DataFrame(
            {
              "a" : [4, 5, 6, 6],
              "b" : [7, 8, 9, 9],
              "c" : [10, 11, 12, 12]
            },
            index = pd.MultiIndex.from_tuples(
              [('d', 1),('d', 2),('e', 2),('e', 3)], 
              names=['n', 'v'])
            )
df
```

```python
# 중복된 데이터를 제거
df = df.drop_duplicates()

df
```
