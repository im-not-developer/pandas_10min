```python
import pandas as pd
import seaborn as sns #데이터 시각화 툴
```

```python
df = sns.load_dataset('iris')

df
```

```python
df.head()
```

```python
columns = ['sepal_width', 'sepal_length', 'species']

df[columns].head()
```

```python
# 특정컬럼만 가져올 수 있다.
df['sepal_width']
```

```python
# 아래와 같이 할 경우 한글 같은 값을 불러오기 어려우니 대괄호를 이용
df.sepal_width
```

```python
# 정규식으로 컬럼을 필터링이 된다.
# _바가 들어간 컬럼을 가져오고 싶다.
df.filter(regex='\_')
```

```python
# length 로 끝나는 컬럼만 가져오고 싶다.
df.filter(regex='length$')
```

```python
# 특정 컬럼으로 시작되는 값을 가져오고 싶다.
df.filter(regex="^sepal")
```

```python
# se로시작하는것
df.filter(regex="^se")
```

```python
df.filter(regex='es$')
```

```python
"""
^(carrot)은 not으로도 사용을 한다.
그래서 species 을 제외하고 나머지 컬럼을 가져온다.
"""
df.filter(regex='^(?!species$).*')
```

```python
df.loc[:,'sepal_width':'petal_width']
```

```python
# 범위를 지정해서 컬럼에 대한 값을 지정을 할 수 있다.
# 행 , 열 배치 인덱스 번호를 포함
df.loc[2:5,'sepal_width':'petal_width']
```

```python
"""
loc와 iloc의 차이는 
loc[2:5] 는 해당 인덱스 즉 5를 포함된 인덱스를 가져온다.

iloc[2:5]은 해당 인덱스 전 값까지 즉 4까지 값을 가져온다.
:은 전체 행

"""
df.iloc[2:5 ,[1,2,4]]
```

```python
#  ,comma를 기준으로 행과 열을 나눌 수 있다.
#     [        행       ]     ,  [                열           ] 
df.loc[df['sepal_width'] > 10, ['sepal_length','sepal_width']]
```
