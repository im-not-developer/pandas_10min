# Subset Observations
## **Logic in Python -and, or, not, xor, any, all 연산 이해하기**


```python
import pandas as pd
```

```python
# 중복된 값들을 추가
df = pd.DataFrame(
            {
              "a" : [4, 5, 6, 6],
              "b" : [7, 8, 9, 9],
              "c" : [10, 11, 12, 12]
            },
            index = pd.MultiIndex.from_tuples(
              [('d', 1),('d', 2),('e', 2),('e', 3)], 
              names=['n', 'v'])
            )
df
```

```python
df = df.drop_duplicates(keep="last")
df
```

```python
df[df["b"] != 7]
```

```python
"""
TypeError: only list-like objects are allowed to be passed to isin(), you passed a [int]
리스트만 받을 수 있다.
"""
df['a'].isin([5, 6])
```

```python
import numpy as np
df = pd.DataFrame(
            {
              "a" : [4, 5, 6, 6, np.nan],
              "b" : [7, 8, np.nan, 9, 9],
              "c" : [10, 11, 12,np.nan, 12]
            },
            index = pd.MultiIndex.from_tuples(
              [('d', 1),('d', 2),('e', 2),('e', 3),('e', 4)], 
              names=['n', 'v'])
            )
df
```

```python
pd.isnull(df)
```

```python
df['a'].isnull().sum()
```

```python
pd.notnull(df)
```

```python
df.notnull().sum()
```

```python
# df.a.notnull()
~df.a.notnull()
```
