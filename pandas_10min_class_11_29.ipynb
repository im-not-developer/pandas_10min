# Reshaping Data

* sort_values, rename, sort_index, reset_index로 데이터 프레임 Reshaping하기

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = sns.load_dataset("mpg")
df.shape
```

```python
df.sort_values('mpg')
```

```python
df.sort_values('cylinders')
```

```python
df.sort_values('mpg').head()
```

```python
# ascending 오르는 상승적인, 위를 향한 [형용사]
# 오름차순
df.sort_values('mpg', ascending=False)
```

```python
# 역순
df.sort_values('mpg', ascending=True).head()
```
