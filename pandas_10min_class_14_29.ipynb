# Combine Data Sets
**merge로 데이터프레임 합치기 left, right, inner, outer 옵션 사용하기**

```python
import pandas as pd
import numpy as np
```

```python
adf = pd.DataFrame({"x1":["A","B","C"], "x2":[1,2,3]})
adf
```

```python
bdf = pd.DataFrame({"x1":["A","B","D"], "x3":["T","F","T"]})
bdf
```

```python
pd.merge(adf, bdf,
         how='left', on='x1')
```

```python
pd.merge(adf, bdf,
         how='right', on='x1')
```

```python
pd.merge(adf, bdf,
         how='inner', on='x1')
```

```python
pd.merge(adf, bdf,
         how='outer', on='x1')
```

```python
bdf.x1.isin(bdf.x1)
```

```python
adf[adf.x1.isin(bdf.x1)]
```

```python
# ~은 값을 제외
adf[~adf.x1.isin(bdf.x1)]
```

```python
ydf = pd.DataFrame({"x1":["A","B","C"], "x2":[1,2,3]})
ydf
```

```python
zdf = pd.DataFrame({"x1":["B","C","D"], "x2":[2,3,4]})
zdf
```

```python
# 기본값은 inner join
pd.merge(ydf, zdf,
         how='outer', on='x1')
```

```python
pd.merge(ydf, zdf)
```

```python
pd.merge(ydf, zdf, how="outer", indicator=True)
```

```python
pd.merge(ydf, zdf, how="outer", indicator=True).query('_merge == "left_only"')
```

```python
pd.merge(ydf, zdf, how="outer", indicator=True).query('_merge == "left_only"').drop(columns=['_merge'])
```
