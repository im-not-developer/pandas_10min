# Summarize Data 
## **apply(function)**

```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df = sns.load_dataset('iris')

df
```

```python
# 익명함수를 통해서 값을 가져올 수 있다.
#lambda x : x[:3] 익명함수로 함수를 만들어서 사용할 수 있다.
df['species'].apply(lambda x : x[:3])
```

```python
df['species_3'] = df['species'].apply(lambda x:x[:3])
df
```

```python
# 뒤에서 3번째까지의 문자를 가져오는 함수
def smp(x):
  x = x[-3:]
  return x
```

```python
df['species-3'] = df['species'].apply(smp)
df
```
