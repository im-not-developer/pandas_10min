# Reshaping Data

**melt, pivot으로 Tidy Data 만들기**


```python
import pandas as pd
import numpy as np
```

```python
 df = pd.DataFrame({'A': {0: 'a', 1: 'b', 2: 'c'},
                    'B': {0: 1, 1: 3, 2: 5},
                    'C': {0: 2, 1: 4, 2: 6}})
 df
```

```python
pd.melt(df, id_vars=['A'], value_vars=['B','C'])
```

```python
# .rename(columns={'variable': 'var', 'value':'val'}) 메소드체이닝
pd.melt(df, value_vars=['A','B','C']).rename(columns={'variable': 'var', 'value':'val'})
```

```python
df2 = pd.DataFrame({'foo': ['one', 'one', 'one', 'two', 'two',
                            'two'],
                    'bar': ['A', 'B', 'C', 'A', 'B', 'C'],
                    'baz': [1, 2, 3, 4, 5, 6],
                    'zoo': ['x', 'y', 'z', 'q', 'w', 't']})

df2
```

```python
df2.pivot(index='foo', columns='bar', values='baz')
```

```python
# 인덱스를 새로 만듬
df3 = df2.pivot(index='foo', columns='bar', values='baz').reset_index()
df3
```

```python
df3.melt(id_vars=['foo'], value_vars=['A','B','C'])
```

```python
df3.melt(id_vars=['foo'], 
         value_vars=['A','B','C']).sort_values(
             ['foo','bar']).rename(columns={'value':'baz'})
```
