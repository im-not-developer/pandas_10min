```python
df = pd.DataFrame(
    {"a": [4,5,6],
     "b": [7,8,9],
     "c": [10,11,12],
    },
    index=[1,2,3]
)
```

```python
df
```

```python
# [행, 열]의 순
df.loc[3, "a"]
```

```python

# [행][열]의 순을 인덱싱해서 가져옴
df.loc[[1,2], ["a","b"]]
```

```python
"""
  index 1,2,3 추출
  columns a, b, c 를 지정을 함
"""

df = pd.DataFrame(
      [[4, 7, 10],
       [5, 8, 11],
       [6, 9, 12]],
      index=[1, 2, 3],
      columns=['a', 'b', 'c'])
df
```

```python
df = pd.DataFrame(
            {
              "a" : [4, 5, 6],
              "b" : [7, 8, 9],
              "c" : [10, 11, 12]
            },
            index = pd.MultiIndex.from_tuples(
              [('d', 1),('d', 2),('e', 2)], 
              names=['n', 'v'])
            )
df
```
